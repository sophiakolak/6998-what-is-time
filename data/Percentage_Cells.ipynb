```python
import numpy as np 
import matplotlib.pyplot as plt  
  
   
#creating the dataset 

data = {'Day 0':2.74,  'Day 5':15.87} 
Day_of_Training = list(data.keys()) 
Percentage_of_Cells = list(data.values()) 
   
fig = plt.figure(figsize = (8, 6)) 
  #calculate %
# creating the bar plot 
plt.bar(Day_of_Training, Percentage_of_Cells, color=['blue', 'green'],  
        width = 0.3) 
  
plt.xlabel("Day of Training") 
plt.ylabel("% of cells that are time cells") 
plt.title("Percentage of time cells before and after learning ") 
plt.show() 

```
